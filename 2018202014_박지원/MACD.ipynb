```python
df_name=['거래날짜','종가']
list
```

```python
df=pd.DataFrame(list,columns=df_name)
df
```

```python
df['이동평균_12일']=0
df['이동평균_26일']=0
df
```

```python
#12일 단위 이동평균
sum_12=0
for i in range(len(df)-26):
    for j in range(12):
        sum_12=sum_12+int(df.iloc[i+j,1])
    df.iloc[i,2]=sum_12/12
    sum_12=0
    
```

```python
#26일 단위 이동평균
sum_26=0
for i in range(len(df)-26):
    for j in range(26):
        sum_26=sum_26+int(df.iloc[i+j,1])
    df.iloc[i,3]=sum_26/26
    sum_26=0
```

```python
df
```

```python
df['MACD']=df['이동평균_12일']-df['이동평균_26일']
df
```

```python
df['시그널']=0
df
```

```python
sum_sg=0
for i in range(len(df)-26):
    for j in range(9):
        sum_sg=sum_sg+int(df.iloc[i+j,4])
    df.iloc[i,5]=sum_sg/9
    sum_sg=0
```

```python
df['오실레이터']=df['MACD']-df['시그널']
```
