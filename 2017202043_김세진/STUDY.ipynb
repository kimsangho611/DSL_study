```python
condition = [
    (data2['거래량등락률'] >= 1) ,
    (data2['거래량등락률'] >= -1) & (data2['거래량등락률'] <1),
    (data2['거래량등락률'] < -1)]
choice = ['상승', '보합', '하락']
data2['전날대비거래량상태'] = np.select(condition, choice, default='Not Specified')
```

```python
for i in range(20,len(data)):
    up_sum=0
    still_sum=0
    down_sum=0
    for j in range(0,20):
        if data.iloc[i-j]['전날대비거래량상태']=='상승':
            up_sum+=data.iloc[i-j]['거래량']
        elif data.iloc[i-j]['전날대비거래량상태']=='보합':
            still_sum+=data.iloc[i-j]['거래량']
        elif data.iloc[i-j]['전날대비거래량상태']=='하락':
            down_sum+=data.iloc[i-j]['거래량']
    vr=(up_sum+still_sum)/(down_sum+still_sum)*100
    data.loc[i,'VR지표']=vr
            
```

```python
def buy_sell(fr,obv,obv_sig,vr,price, date):
    sig_Buy=[]
    sig_Sell=[]
    
    flag=0 #to check buy, if flag is 10 can not buy
    money=100000000
    cur_money=money
    count=0 # 가지고 있는 주식 수
    state=0 #과열상태 확인 state
    A=3# 현재 금액의 A 퍼센트 만큼 매수
    tax=0.0008 #세금
    charge=0.002 #수수료
    start = 100
    last_buy_max=0
    state='침착' #침착, 과매도 ,보통, 과매수 과열 천정
    if(fr[vr][start]<=75):
        state='침착'
    elif(fr[vr][start]<=100):
        state='과매도'
    elif(fr[vr][start]<=150):
        state='보통'
    elif(fr[vr][start]<=300):
        state='보통'
    elif(fr[vr][start]<=450):
        state='과열'
    else :
        state='천정'
    for i in range(0,start):
        sig_Buy.append(np.nan)
        sig_Sell.append(np.nan)
    for i in range(start,len(fr)):
        if (fr[obv][i]>fr[obv_sig][i] and flag<A) and ((state=='침착' and fr[vr][i]>=75) or 300<=fr[vr][i]<=450) : #매수 
            count=count+(cur_money/A)/fr[price][i]
            cur_money=cur_money-(cur_money/A)*(1-tax-charge)
            flag+=1
            if(last_buy_max<fr[price][i]):
                last_buy_max=fr[price][i]
            print("*****매수 정보 : ",fr[date][i],"   ",fr[price][i],"원 매수******")
            sig_Buy.append(fr[price][i])
            sig_Sell.append(np.nan)
        elif (flag>0 and fr[price][i]<last_buy_max*0.95)or ((fr[obv][i-1]<fr[obv_sig][i-1] and flag>0) and ((state =='천정' or (state !='침착' and fr [vr][i]<=75)) or (state =='과열' and (fr[vr][i]<=300 or fr [vr][i]>=450)))): #매도 
            cur_money+=fr[price][i]*(count/flag)*(1-charge)
            count=count-(count/flag)
            flag-=1
            sig_Sell.append(fr[price][i])
            sig_Buy.append(np.nan)
            if(fr[price][i]<last_buy_max*0.95):
                last_buy_max=0
            print("*****매도 정보 : ",fr[date][i],"   ",fr[price][i],"원 매도******")
        else:
            sig_Buy.append(np.nan)
            sig_Sell.append(np.nan)
        if(fr[vr][i]<=75):
            state='침착'
        elif(fr[vr][i]<=100):
            state='과매도'
        elif(fr[vr][i]<=150):
            state='보통'
        elif(fr[vr][i]<=300):
            state='보통'
        elif(fr[vr][i]<=450):
            state='과열'
        else :
            state='천정'
    total=cur_money+count*fr[price][len(fr)-1]
    print("최종 금액 : ",total)
    return (sig_Buy,sig_Sell)
```

```python
x=buy_sell(data,'OBV지표','OBV_SIG','VR지표','거래가', '날짜')
data['매수']=x[0]
data['매도']=x[1]
data
```
