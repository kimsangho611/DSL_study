    딕셔너리는 말 그대로 사전과 같은 형태의 데이터 구조이다.
    사전의 단어는 딕셔너리의 key
    단어의 설명은 딕셔너리의 value값이라고 칭한다. 쌍으로 저장 가능

## **3. 튜플(tuple)**
    데이터가 고정이 되어 변경이 불가능 한 데이터 집합
    자료구조로는 잘 사용하지 않는다.
    ( ) 소괄호로 묶어 사용한다.
    구분은 , 콤마로 사용한다.

### 3-1. tuple 생성

```python
test_tuple = () # 의미가 모호해서 잘 사용하지 않는다.
test_tuple = tuple()
test_tuple = 1, 2, 3, 4
```

### 3-2. tuple 인덱싱

```python
test_tuple[0]
```

### 3-3. tuple 업데이트 (안됨)

```python
test_tuple[0] = 0
```

### 3-4. tuple 값 추가

```python
test_tuple + (7, )
```
