# Pandas의 DataFrame, Series operation

## Series operation

```python
s1 = Series(range(1,6),index = list("abcde"))
s2 = Series(range(6,11),index = list("cdefg"))
```

```python
s1.add(s2)
```

```python
s1 + s2
```

```python
s1.add(s2,fill_value = 0) 
# fill_value = 0을 쓰면 인덱스가 없는 Series에는 value에 0을 할당
```

## DataFrame operation
   - DataFrame은 column과 index 모두 고려
   - add operation 을 쓰면 NaN값 변환 시킬수 있음
   - Operation type: add, sub, div, mul

```python
df1 = DataFrame(np.arange(4).reshape(2,2),columns = list("ab"))
df2 = DataFrame(np.arange(9).reshape(3,3),columns = list("abc"))
```

```python
df1 + df2
```

```python
df1.add(df2,fill_value = 0)
```

## Series + DataFrame
   - pandas의 broadcasting
   - 컬럼 인덱스 기준으로 연산함

```python
df = DataFrame(np.arange(9).reshape(3,3),columns = list("abc"))
s = Series(np.arange(10,13),index = list("abc"))
s2 = Series(np.arange(10,13))
```

```python
df + s  # 같은 인덱스가 존재하므로 연산됨
```

```python
df + s2
```

```python
df.add(s2,axis = 0) # axis를 사용해서 column기준을 row기준으로 바꿈
```

# Pandas의 Lambda함수(익명 함수)
   - 한 줄로 함수를 표현하는 익명 함수 기법
   - Lisp 언어에서 시작된 기법으로 오늘날 현대 언어에 많이 사용

```python
def f(x): # python에서 함수 사용
    return x + 5
f(3)
```

```python
f = lambda x:x+5 # 위의 함수를 간단하게 표현
f(3)
```

```python
f = lambda x: x / 3 # 하나의 인수만 처리
f(6)
```

```python
f = lambda x: x ** 2 # 하나의 인수만 처리
f(4)
```

```python
(lambda x: x+4)(2) # 이름을 할당하지 않는 lambda
```

# Pandas의 map() 함수
   - 형식: map( function ,  sequence )
   - 실제 값을 보기 위해서는 list,tuple함수를 사용(아니면 주소값을 보여줌)

```python
seq = [1,2,3,4,5]
f = lambda x : x * 3
list(map(f,seq))
```

```python
f = lambda x,y:x+y # 두개 이상의 인수가 있으면 두개의 sequece형을 써야됨
list(map(f,seq,seq))
```

```python
list(map(lambda x: x*x,seq)) # 익명함수를 바로 사용
# python3에서 list함수를 쓰지 않으면 memory주소가 나옴
```

## map for Series
   - Pandas의 series type의 데이터에도 map함수 사용가능
   - function 대신 dict, sequence형 자료등으로 대체 가능

```python
s = Series(np.arange(10))
s.head(5)
```

```python
s.map(lambda x:x + 3).head(5)
```

```python
dic = {1:'apple',2:'banana',3:'carrot'}
s.map(dic).head(5) 
# dict type으로 데이터 교체, 값이 없으면 NaN
# dict를 넣으면 키 값이 인덱스로 들어감
```

```python
s2 = Series(np.arange(100,500,100))
s.map(s2).head(5)
```

# Pandas의 apply(), applymap()

```python
df1_1 = df1
df1_1.drop([18],axis = 0,inplace = True)
df1_1.tail()
```

```python
df1_2 = df1_1[["휘발유","경유"]]
df1_2 = df1_2.astype(float)
df1_2.head()
```

## apply()
   - Series or DataFrame 전체 컬럼 전체에 함수를 적용함
   - 특정 컬럼에만 해줌

```python
f = lambda x:x.max() - x.min()
df1_2.apply(f) # 각 column별로 결과 값 반환
```

```python
df1_2.sum()
```

```python
df1_2.apply(sum)
```

```python
def f(x):
    return Series([x.min(),x.max(),x.mean()],index = ["min","max","mean"])
df1_2.apply(f)
```

## applymap()
   - applymap()은 컬럼 전체에 대한 연산을 한다.
   - map()는 Series에서만 쓰지만 applymap()은 dataframe에서 쓴다.

```python
f = lambda x: -x
df1_2.applymap(f).head(5)
```

```python
f = lambda x:-x
df1_2["휘발유"].apply(f).head()
```

## Correlation & Covariance
   - 상관 계수와 공분산을 구하는 함수
   - corr, cov, corrwith
   > 기준: 
           30% 미만 -> 상관이 거의 없다.
           70% 미만 -> 상관이 어느 정도 있다.
           70% 이상 -> 상관이 매우 있다

```python
df["살인 발생"].corr(df["강도 발생"]) # 두 컬럼의 상관관계 계산
```

```python
df["살인 발생"].cov(df["강도 발생"]) # 변수간의 공분산 계산
```

```python
df.corrwith(df["강도 발생"]) 
# 다른 Series나 DataFrame과 상관관계계산
# 하나의 변수와 나머지 변수 간의 상관관계를 볼때 좋음
```

```python
df.corr() # 전체 상관관계
```
